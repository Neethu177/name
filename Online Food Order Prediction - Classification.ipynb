```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
import pandas as pd
s=pd.read_csv("onlinedeliverydata (1).csv")
s
```

```python
s.info()
```

```python
s.describe()
```

```python
s=s.drop(['Politeness','Medium (P1)','Medium (P2)','Meal(P1)','Meal(P2)','Perference(P1)','Perference(P2)','Bad past experience','Temperature','Number of calls','Delivery person ability','Ease and convenient','Self Cooking','Health Concern','Number of calls','Late Delivery','Time saving','Maximum wait time','Unaffordable','Delay of delivery person getting assigned','Delay of delivery person picking up food','Unavailability','Long delivery time','Low quantity low time','Good Road Condition','Wrong order delivered','Missing item','Residence in busy location','Google Maps Accuracy','Good Tracking system'],axis=1)
s
```

```python
s.dropna(inplace=True)
```

```python
s.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
s["Easy Payment option "]=le.fit_transform (s["Easy Payment option"])
s["More Offers and Discount"]=le.fit_transform(s["More Offers and Discount"])
s["Good Food quality"]=le.fit_transform(s["Good Food quality"])
s["Order Time"]=le.fit_transform (s["Order Time"])
s["High Quality of package"]=le.fit_transform(s["High Quality of package"])
s["Output"]=le.fit_transform(s["Output"])
s["Reviews"]=le.fit_transform(s["Reviews"])
s["More restaurant choices"]=le.fit_transform(s["More restaurant choices"])
s["Influence of rating"]=le.fit_transform(s["Influence of rating"])
s["Influence of time"]=le.fit_transform(s["Influence of time"])
s["Less Delivery time"]=le.fit_transform(s["Less Delivery time"])
s["Good Quantity"]=le.fit_transform(s["Good Quantity"])
s["Gender"]=le.fit_transform(s["Gender"])
s["Educational Qualifications"]=le.fit_transform(s["Educational Qualifications"])
s["Marital Status"]=le.fit_transform(s["Marital Status"])
s["Occupation"]=le.fit_transform(s["Occupation"])
s["Monthly Income"]=le.fit_transform(s["Monthly Income"])
s


```

```python
x=np.array(s[["Gender","Age","Marital Status","Occupation","Monthly Income","Educational Qualifications","Family size","Pin code"]])
x
```

```python
y=np.array(s[["Output"]])
y
```

```python
from matplotlib import pyplot as plt
plt.hist(s["Age"],rwidth=0.80)
plt.title("Histogram")
plt.xlabel("Age")
plt.ylabel("Output")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(random_state=0)
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn import metrics
LR=metrics.accuracy_score(y_test,y_pred)
LR
```

```python
from sklearn.tree import DecisionTreeClassifier
dc=DecisionTreeClassifier()
dc.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn import metrics
DT=metrics.accuracy_score(y_test,y_pred)
DT
```

```python
from sklearn.ensemble import RandomForestClassifier
lr=RandomForestClassifier(n_estimators=3,random_state=0)
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn import metrics
RF=metrics.accuracy_score(y_test,y_pred)
RF
```

```python
from sklearn.svm import SVC
dc=SVC(kernel='linear')
dc.fit(x_train,y_train)
```

```python
y_pred=dc.predict(x_test)
y_pred
```

```python
from sklearn import metrics
SR=metrics.accuracy_score(y_test,y_pred)
SR
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNC=KNeighborsClassifier(n_neighbors=5)
```

```python
KNC=KNC.fit(x_train,y_train)
KNC
```

```python
y_pred=KNC.predict(x_test)
y_pred
```

```python
from sklearn import metrics
KN=metrics.accuracy_score(y_pred,y_test)
KN
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.naive_bayes import GaussianNB
GN=GaussianNB()
GN.fit(x_train,y_train)
y_pred=GN.predict(x_test)
y_pred
```

```python
from sklearn import metrics
NB=metrics.accuracy_score(y_test,y_pred)
NB
```

```python
model=pd.DataFrame({
    'model':['logisticRegression','DecisionTreeClassifier','RandomForestClassifier','svr','kn','NB'],
    'accuracyscore':[LR,DT,RF,SR,KN,NB]
})
model
sns.barplot(x='accuracyscore',y='model',data=model)
model.sort_values(by='accuracyscore',ascending=False)
```
